```python
from scipy.stats import pearsonr, spearmanr
#Calculate correlations
pearson_mean_pca, _ = pearsonr(m_albedo_mean, m_albedo_pca_n20)
spearman_mean_pca, _ = spearmanr(m_albedo_mean, m_albedo_pca_n20)

pearson_mean_kpca, _ = pearsonr(m_albedo_mean, m_albedo_kpca)
spearman_mean_kpca, _ = spearmanr(m_albedo_mean, m_albedo_kpca)

pearson_pca_kpca, _ = pearsonr(m_albedo_pca_n20, m_albedo_kpca)
spearman_pca_kpca, _ = spearmanr(m_albedo_pca_n20, m_albedo_kpca)

# Plot comparisons
plt.figure(figsize=(15, 5))

# Mean vs PCA
plt.subplot(1, 3, 1)
plt.scatter(m_albedo_mean, m_albedo_pca_n20, alpha=0.5)
plt.title(f"Mean vs PCA (n=20)\nPearson: {pearson_mean_pca:.2f}, Spearman: {spearman_mean_pca:.2f}")
plt.xlabel("SIF - mean - mW m⁻² nm⁻¹ sr⁻¹")
plt.ylabel("SIF - PCA (n=20) - mW m⁻² nm⁻¹ sr⁻¹")

# Mean vs Kernel PCA
plt.subplot(1, 3, 2)
plt.scatter(m_albedo_mean, m_albedo_kpca, alpha=0.5)
plt.title(f"Mean vs Kernel PCA\nPearson: {pearson_mean_kpca:.2f}, Spearman: {spearman_mean_kpca:.2f}")
plt.xlabel("SIF - mean - mW m⁻² nm⁻¹ sr⁻¹")
plt.ylabel("SIF - KPCA - mW m⁻² nm⁻¹ sr⁻¹")

# PCA vs Kernel PCA
plt.subplot(1, 3, 3)
plt.scatter(m_albedo_pca_n20, m_albedo_kpca, alpha=0.5)
plt.title(f"PCA (n=20) vs Kernel PCA\nPearson: {pearson_pca_kpca:.2f}, Spearman: {spearman_pca_kpca:.2f}")
plt.xlabel("SIF - PCA (n=20) - mW m⁻² nm⁻¹ sr⁻¹")
plt.ylabel("SIF - KPCA - mW m⁻² nm⁻¹ sr⁻¹")

# PCA (n=10) vs PCA (n=20)
pearson_pca10_pca20, _ = pearsonr(m_albedo_pca_n10, m_albedo_pca_n20)
spearman_pca10_pca20, _ = spearmanr(m_albedo_pca_n10, m_albedo_pca_n20)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.scatter(m_albedo_pca_n10, m_albedo_pca_n20, alpha=0.5)
plt.title(f"PCA (n=10) vs PCA (n=20)\nPearson: {pearson_pca10_pca20:.2f}, Spearman: {spearman_pca10_pca20:.2f}")
plt.xlabel("SIF - PCA (n=10) - mW m⁻² nm⁻¹ sr⁻¹")
plt.ylabel("SIF - PCA (n=20) - mW m⁻² nm⁻¹ sr⁻¹")

# PCA vs PCA (n=20)
pearson_pca_pca20, _ = pearsonr(m_albedo_pca_3, m_albedo_pca_n20)
spearman_pca_pca20, _ = spearmanr(m_albedo_pca_3, m_albedo_pca_n20)

plt.subplot(1, 2, 2)
plt.scatter(m_albedo_pca_3, m_albedo_pca_n20, alpha=0.5)
plt.title(f"PCA (n=50) vs PCA (n=20)\nPearson: {pearson_pca_pca20:.2f}, Spearman: {spearman_pca_pca20:.2f}")
plt.xlabel("SIF - PCA (n=50) - mW m⁻² nm⁻¹ sr⁻¹")
plt.ylabel("SIF - PCA (n=20) - mW m⁻² nm⁻¹ sr⁻¹")

plt.tight_layout()
plt.show()
```

```python
# Calculate pixel-wise statistics
sif_values = np.stack([fix_albedo_mean, fix_albedo_pca, m_albedo_kpca], axis=0)
std_map = np.std(sif_values, axis=0)  # Standard deviation
range_map = np.ptp(sif_values, axis=0)  # Range (max - min)
mean_map = np.mean(sif_values, axis=0)  # Mean
coeff_var_map = std_map / mean_map  # Coefficient of variation (std/mean)

# Plot the maps
plt.figure(figsize=(15, 10))

# Standard deviation map
plt.subplot(2, 2, 1)
plt.plot(std_map, label="Standard Deviation")
plt.title("Pixel-wise Standard Deviation of SIF")
plt.xlabel("Pixel Index")
plt.ylabel("Standard Deviation")
plt.legend()

# Range map
plt.subplot(2, 2, 2)
plt.plot(range_map, label="Range", color="orange")
plt.title("Pixel-wise Range of SIF")
plt.xlabel("Pixel Index")
plt.ylabel("Range")
plt.legend()

# Coefficient of variation map
plt.subplot(2, 2, 3)
plt.plot(coeff_var_map, label="Coefficient of Variation", color="green")
plt.title("Pixel-wise Coefficient of Variation of SIF")
plt.xlabel("Pixel Index")
plt.ylabel("Coefficient of Variation")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Prepare summary list
summary = []

def summarize_sif(values, method, m, region):
    avg = np.mean(values)
    std = np.std(values)
    summary.append({
        "Tau Method": method,
        "Albedo Method": m,
        "Region": region,
        "Avg SIF in mW m⁻² nm⁻¹": round(avg, 4),
        "Std Dev": round(std, 4),
        #"Notes": note
    })

#print(np.mean(m_albedo_kpca_2))
# Add entries (customize as needed)
summarize_sif(m_albedo_mean, "Mean", "Modeled albedo s=5, m=3", "Sahara")
summarize_sif(m_albedo_mean_2, "Mean", "Modeled albedo s=5, m=2", "Sahara")
summarize_sif(m_albedo_mean_4, "Mean", "Modeled albedo s=5, m=4", "Sahara")
summarize_sif(m_albedo_mean_5, "Mean", "Modeled albedo s=5, m=5", "Sahara")
summarize_sif(fix_albedo_mean, "Mean", "Fixed albedo", "Sahara")
summarize_sif(fix_albedo_pca, "PCA", "Fixed albedo", "Sahara")
summarize_sif(m_albedo_pca_2, "PCA", "Modeled albedo s=5, m=2", "Sahara")
summarize_sif(m_albedo_pca_4, "PCA", "Modeled albedo s=5, m=4", "Sahara")
summarize_sif(m_albedo_pca_5, "PCA", "Modeled albedo s=5, m=5", "Sahara")
summarize_sif(m_albedo_pca_n10, "PCA-n=10", "Modeled albedo s=5, m=3", "Sahara")
summarize_sif(m_albedo_pca, "PCA-n=20", "Modeled albedo s=5, m=3", "Sahara")
summarize_sif(m_albedo_pca_3, "PCA-n=50", "Modeled albedo s=5, m=3", "Sahara")
summarize_sif(fix_albedo_kpca, "KPCA", "Fixed albedo" , "Sahara")
summarize_sif(m_albedo_kpca_a, "KPCA", "Modeled albedo s=5, m=3", "Sahara A")
#summarize_sif(data_kpca_m2["alb_2"], "KPCA", "Modeled albedo s=2, m=3", "Sahara", "calculate again")
summarize_sif(m_albedo_kpca_2, "KPCA", "Modeled albedo s=5, m=2", "Sahara")
summarize_sif(m_albedo_kpca_3, "KPCA", "Modeled albedo s=5, m=3", "Sahara")
summarize_sif(m_albedo_kpca_4, "KPCA", "Modeled albedo s=5, m=4", "Sahara")
summarize_sif(m_albedo_kpca_5, "KPCA", "Modeled albedo s=5, m=5", "Sahara")

# Create dataframe
df_summary = pd.DataFrame(summary)

# Display or save
print(df_summary)
df_summary.to_csv("sif_summary.csv", index=False)  # optional
latex_code = df_summary.to_latex(index=False, float_format="%.3f")
print(latex_code)
```
