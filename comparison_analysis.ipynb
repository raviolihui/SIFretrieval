```python
summary = []

def summarize_sif(values, method, m, region):
    avg = np.mean(values)
    std = np.std(values)
    summary.append({
        "Tau Method": method,
        "Albedo Method": m,
        "Region": region,
        "Avg SIF in mW m nm": round(avg, 4),
        "Std Dev": round(std, 4),
    })


summarize_sif(m_albedo_mean_2, "Mean", "Modeled albedo, m=2", "Sahara")
summarize_sif(m_albedo_mean_3, "Mean", "Modeled albedo, m=3", "Sahara")
summarize_sif(m_albedo_mean_4, "Mean", "Modeled albedo, m=4", "Sahara")
summarize_sif(m_albedo_mean_1, "Mean", "Modeled albedo, m=1", "Sahara")
summarize_sif(fix_albedo_mean, "Mean", "Fixed albedo", "Sahara")
summarize_sif(fix_albedo_pca_n10, "PCA", "Fixed albedo", "Sahara")
summarize_sif(m_albedo_pca_n20_1, "PCA-n=20", "Modeled albedo, m=1", "Sahara")
summarize_sif(m_albedo_pca_n20_2, "PCA-n=20", "Modeled albedo, m=2", "Sahara")
summarize_sif(m_albedo_pca_n20_3, "PCA-n=20", "Modeled albedo, m=3", "Sahara")
summarize_sif(m_albedo_pca_n20_4, "PCA-n=20", "Modeled albedo, m=4", "Sahara")
summarize_sif(m_albedo_pca_n10_1, "PCA-n=10", "Modeled albedo, m=1", "Sahara")
summarize_sif(m_albedo_pca_n10_2, "PCA-n=10", "Modeled albedo, m=2", "Sahara")
summarize_sif(m_albedo_pca_n10_3, "PCA-n=10", "Modeled albedo, m=3", "Sahara")
summarize_sif(m_albedo_pca_n10_4, "PCA-n=10", "Modeled albedo, m=4", "Sahara")
summarize_sif(m_albedo_pca_n10, "PCA-n=10", "Modeled albedo, m=2", "Sahara")
summarize_sif(m_albedo_pca_n30, "PCA-n=30", "Modeled albedo, m=2", "Sahara")
summarize_sif(m_albedo_pca_n40, "PCA-n=40", "Modeled albedo, m=2", "Sahara")
summarize_sif(m_albedo_pca_n50, "PCA-n=50", "Modeled albedo, m=2", "Sahara")
summarize_sif(fix_albedo_kpca, "KPCA", "Fixed albedo" , "Sahara")
summarize_sif(m_albedo_kpca_d1, "KPCA", "Modeled albedo, m=3", "River Chad")
summarize_sif(fix_albedo_kpca_d1, "KPCA", "Fixed albedo, Sahara a", "River Chad")
summarize_sif(m_albedo_kpca_2, "KPCA", "Modeled albedo, m=2", "Sahara")
summarize_sif(m_albedo_kpca_3, "KPCA", "Modeled albedo, m=3", "Sahara")
summarize_sif(m_albedo_kpca_4, "KPCA", "Modeled albedo, m=4", "Sahara")
summarize_sif(m_albedo_kpca_1, "KPCA", "Modeled albedo, m=1", "Sahara")

df_summary = pd.DataFrame(summary)

print(df_summary)
#df_summary.to_csv("sif_summary.csv", index=False)  
latex_code = df_summary.to_latex(index=False, float_format="%.3f")
print(latex_code)
```
