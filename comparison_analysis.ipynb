```python
# Calculate pixel-wise statistics
sif_values = np.stack([fix_albedo_mean, fix_albedo_pca, m_albedo_kpca], axis=0)
std_map = np.std(sif_values, axis=0)  # Standard deviation
range_map = np.ptp(sif_values, axis=0)  # Range (max - min)
mean_map = np.mean(sif_values, axis=0)  # Mean
coeff_var_map = std_map / mean_map  # Coefficient of variation (std/mean)

# Plot the maps
plt.figure(figsize=(15, 10))

# Standard deviation map
plt.subplot(2, 2, 1)
plt.plot(std_map, label="Standard Deviation")
plt.title("Pixel-wise Standard Deviation of SIF")
plt.xlabel("Pixel Index")
plt.ylabel("Standard Deviation")
plt.legend()

# Range map
plt.subplot(2, 2, 2)
plt.plot(range_map, label="Range", color="orange")
plt.title("Pixel-wise Range of SIF")
plt.xlabel("Pixel Index")
plt.ylabel("Range")
plt.legend()

# Coefficient of variation map
plt.subplot(2, 2, 3)
plt.plot(coeff_var_map, label="Coefficient of Variation", color="green")
plt.title("Pixel-wise Coefficient of Variation of SIF")
plt.xlabel("Pixel Index")
plt.ylabel("Coefficient of Variation")
plt.legend()

plt.tight_layout()
plt.show()
```
